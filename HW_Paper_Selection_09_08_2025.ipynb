The paper I've selected is here: 10.1001/jamanetworkopen.2024.53756 and examines the rate of conversion from mild cognitive impairment (MCI) to Alzheimer's Dementia (AD). They developed an index to predict age at onset of MCI/AD called the FDI. In Figure 2, they plot survival curves for participants being free of MCI and being free of dementia. This is an example of an exponential distribution because it tracks a binomial outcome (has or does not have MCI/Dementia) over a continuous metric (here their FDI score).

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.stats import binom
from scipy.stats import expon
```

```python
# Use code from tutorial for exponential distribution:

delta_t = 1000
plambda = 1
num_bins_per_second = 100
total_bins = round(num_bins_per_second * delta_t )

p = plambda * delta_t / total_bins
outcomes = binom.rvs(1, p, size = total_bins)

intervals_between_events = np.diff(np.where(outcomes == 1))
intervals_between_events_sec = intervals_between_events / num_bins_per_second


counts, edges = np.histogram(intervals_between_events_sec, bins=50)
plt.bar(edges[:-1], np.divide(counts, np.trapz(counts, edges[:-1])), width=0.15, align='edge')
plt.title('Conversion to MCI Modeled as an Exponential Distribution')
plt.xlabel('FDI Index')
plt.ylabel('% Free of MCI')

plt.show()


```

This plot is similar to the survival curves in Figure 2 (though of course the numbers are off). As the FDI index increases, the % of participants free of an MCI diagnosis drops.
